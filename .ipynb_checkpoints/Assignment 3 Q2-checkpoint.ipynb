```r
require(mice)
load("dataex2(1).Rdata")
data <- dataex2
```

```r
simdata <- data[, , 1]
print(simdata)
imps <- mice(simdata, m=20, method = "norm.boot", printFlag = FALSE, seed = 1)
fits <- with(imps, exp = lm(Y ~ X))
```

```r
ests <- pool(fits)

ff <- summary(ests)
```

```r
ff <- summary(ests, conf.int = TRUE)
```

```r
ff$'97.5 %'
```

```r
coverage=0
beta1 = 3
for (i in 1:100){
    simdata <- data[, , 100]
    imps <- mice(simdata, m=20, method = "norm.nob", printFlag = FALSE, seed = 1)
    fits <- with(imps, exp = lm(Y ~ X))
    est <- pool(fits)
    coefs <- summary(ests, conf.int = TRUE)
    # confidence interval, upper limit
    ul <- coefs$'97.5 %'
    if(ul[1] >= beta1){
        coverage = coverage + 1
    }
    
}
```
